```python
from datasets import load_dataset
import networkx as nx
import matplotlib.pyplot as plt 
import numpy as np
import tqdm
import pickle
import os
```

```python
def dfs(adj_matrix, visited, node):
    visited[node] = True
    for i in range(len(adj_matrix)):
        if adj_matrix[node][i] == 1 and not visited[i]:
            dfs(adj_matrix, visited, i)
def is_connected(adj_matrix):
    num_nodes = len(adj_matrix)
    visited = [False] * num_nodes
    dfs(adj_matrix, visited, 0)
    return all(visited)
```

```python
def is_connected(graph):
    G = nx.from_numpy_array(graph)
    return nx.is_connected(G)

dataset_path = 'graph_dataset.pkl'

if os.path.exists(dataset_path):
    with open(dataset_path, 'rb') as f:
        dataset = pickle.load(f)
    print("Veri seti dosyadan yüklendi.")
else:
    dataset = []
    for i in tqdm.tqdm(range(500)):
        graph = nx.to_numpy_array(nx.erdos_renyi_graph(1000, 0.80))
        while not is_connected(graph):
            graph = nx.to_numpy_array(nx.erdos_renyi_graph(1000, 0.80))
        dataset.append(graph)
    
    with open(dataset_path, 'wb') as f:
        pickle.dump(dataset, f)
    print("Veri seti oluşturuldu ve kaydedildi.")

print("Veri setinden bir örnek:", dataset[0])
```

```python
class GraphVisualization: 
    def __init__(self):
        self.visual = [] 
    def addEdge(self, a, b): 
        temp = [a, b] 
        self.visual.append(temp) 
    def visualize(self): 
        G = nx.Graph() 
        G.add_edges_from(self.visual) 
        nx.draw_networkx(G) 
        plt.show() 
```

```python
graphsage_embed = []
for adjacency_matrix in tqdm.tqdm(dataset):
    num_nodes = len(adjacency_matrix)
    embedding_dim = 64
    node_embeddings_graphsage = np.random.randn(num_nodes, embedding_dim)

    def graphsage_update(adj_matrix, node_embeddings, node, dim=64):
        neighbors = np.where(adj_matrix[node] == 1)[0] 
        if len(neighbors) == 0:
            return node_embeddings[node]
        else:
            neighbor_embeddings = node_embeddings[neighbors]
            aggregated_embedding = np.mean(neighbor_embeddings, axis=0)
            return aggregated_embedding

    num_iterations = 10
    for _ in range(num_iterations):
        updated_embeddings = np.zeros_like(node_embeddings_graphsage)
        for i in range(num_nodes):
            updated_embeddings[i] = graphsage_update(adjacency_matrix, node_embeddings_graphsage, i)
        node_embeddings_graphsage = updated_embeddings
    graphsage_embed.append(node_embeddings_graphsage)
```

```python
with open("graphsage_embed.pkl", 'wb') as f:
        pickle.dump(graphsage_embed, f)
```

```python
gcn_embed = []
for adjacency_matrix in tqdm.tqdm(dataset):
    adjacency_matrix = np.array(adjacency_matrix)
    num_nodes = len(adjacency_matrix)
    embedding_dim = 64
    node_embeddings_gcn = np.random.randn(num_nodes, embedding_dim)

    def gcn_propagate(adj_matrix, node_embeddings, dim=64):
        normalized_adj = adj_matrix / adj_matrix.sum(axis=1, keepdims=True)
        updated_embeddings = np.dot(normalized_adj, node_embeddings)
        return updated_embeddings

    num_iterations = 10
    for _ in range(num_iterations):
        node_embeddings_gcn = gcn_propagate(adjacency_matrix, node_embeddings_gcn)

    gcn_embed.append(node_embeddings_gcn)
```

```python
from gensim.models import Word2Vec
node2vec_embed = []
for adjacency_matrix in tqdm.tqdm(dataset):
    graph = adjacency_matrix
    num_walks = 10 
    walk_length = 10
    walks = []

    for node in range(len(graph)):
        for _ in range(num_walks):
            walk = [str(node)]
            current_node = node
            for _ in range(walk_length - 1):
                neighbors = [str(neighbor) for neighbor, edge_exists in enumerate(graph[current_node]) if edge_exists]
                next_node = np.random.choice(neighbors)
                walk.append(next_node)
                current_node = int(next_node)
            walks.append(walk)

    model = Word2Vec(walks, vector_size=64, window=5, min_count=0, sg=1, workers=4, epochs=20)

    node_embeddings = []
    for node in range(len(graph)):
        node_embeddings.append(model.wv[str(node)])
    node2vec_embed.append(node_embeddings)
```

```python
with open("node2vec_embed.pkl", 'wb') as f:
        pickle.dump(node2vec_embed, f)
```

```python
graph_arr = []
for i in tqdm.tqdm(range(len(dataset))):
    graph = dataset[i]
    G = nx.Graph();
    for i in range(len(graph)):
        G.add_node(i)

    for i in range(len(graph)):
        for j in range(len(graph[0])):
            if(graph[i][j] == 1):
                G.add_edge(i, j)
    graph_arr.append(G)
```

```python
from sklearn.model_selection import train_test_split
import random

labels = []
for i in range(len(dataset)):
    labels.append(random.randint(0,1))
for i in range(len(graphsage_embed)):
    graphsage_embed[i] = np.reshape(np.array(graphsage_embed[i]), (-1,)).tolist()
gs_X_train, gs_X_test, gs_y_train, gs_y_test = train_test_split(graphsage_embed, labels, test_size=0.2)
gcn_X_train, gcn_X_test, gcn_y_train, gcn_y_test = train_test_split(gcn_embed, labels, test_size=0.2)
n2v_X_train, n2v_X_test, n2v_y_train, n2v_y_test = train_test_split(node2vec_embed, labels, test_size=0.2)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

gs_model = SVC()
gcn_model = SVC()
n2v_model = SVC()

gs_model.fit(gs_X_train, gs_y_train)
gcn_model.fit(gcn_X_train, gcn_y_train)
n2v_model.fit(n2v_X_train, n2v_y_train)

gs_pred = gs_model.predict(gs_X_test)
gcn_pred = gcn_model.predict(gcn_X_test)
n2v_pred = n2v_model.predict(n2v_X_test)

gs_accuracy = accuracy_score(gs_y_test, gs_pred)
gs_precision = precision_score(gs_y_test, gs_pred)
gs_recall = recall_score(gs_y_test, gs_pred)
gs_f1 = f1_score(gs_y_test, gs_pred)

gcn_accuracy = accuracy_score(gcn_y_test, gcn_pred)
gcn_precision = precision_score(gcn_y_test, gcn_pred)
gcn_recall = recall_score(gcn_y_test, gcn_pred)
gcn_f1 = f1_score(gcn_y_test, gcn_pred)

n2v_accuracy = accuracy_score(n2v_y_test, n2v_pred)
n2v_precision = precision_score(n2v_y_test, n2v_pred)
n2v_recall = recall_score(n2v_y_test, n2v_pred)
n2v_f1 = f1_score(n2v_y_test, n2v_pred)

print("Metrics for gs_model:")
print(f"Accuracy: {gs_accuracy}")
print(f"Precision: {gs_precision}")
print(f"Recall: {gs_recall}")
print(f"F1-score: {gs_f1}")
print()

print("Metrics for gcn_model:")
print(f"Accuracy: {gcn_accuracy}")
print(f"Precision: {gcn_precision}")
print(f"Recall: {gcn_recall}")
print(f"F1-score: {gcn_f1}")
print()

print("Metrics for n2v_model:")
print(f"Accuracy: {n2v_accuracy}")
print(f"Precision: {n2v_precision}")
print(f"Recall: {n2v_recall}")
print(f"F1-score: {n2v_f1}")
```
